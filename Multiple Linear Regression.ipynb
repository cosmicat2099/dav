```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Define the data
x1 = np.array([1, 2, 3, 4, 5, 6])
x2 = np.array([2, 4, 6, 8, 10, 12])
y = np.array([6, 9, 12, 15, 18, 21])

# Fit linear regression model for x1
model_x1 = LinearRegression().fit(x1.reshape(-1, 1), y)
y_pred_x1 = model_x1.predict(x1.reshape(-1, 1))

# Fit linear regression model for x2
model_x2 = LinearRegression().fit(x2.reshape(-1, 1), y)
y_pred_x2 = model_x2.predict(x2.reshape(-1, 1))

# Print estimated coefficients
print(f"Estimated coefficients for x1: {model_x1.coef_}")
print(f"Estimated coefficients for x2: {model_x2.coef_}")

# Plot the data and regression lines for both x1 and x2
fig, ax = plt.subplots()
ax.scatter(x1, y, s=10, label='x1')
ax.scatter(x2, y, s=10, label='x2')
ax.plot(x1, y_pred_x1, color='red', linewidth=2)
ax.plot(x2, y_pred_x2, color='green', linewidth=2)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.show()
```
