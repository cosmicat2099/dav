```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Import the csv file and convert the 'Month' column to a datetime format
air = pd.read_csv('air.csv')
air['Month'] = pd.to_datetime(air['Month'])

# Set the 'Month' column as the index of the dataframe
air.set_index('Month', inplace=True)

# Visualize the time series by plotting it
plt.plot(air)
plt.xlabel('Year')
plt.ylabel('Passengers (in thousands)')
plt.title('Air Passengers Dataset')
plt.show()

# Fit an ARIMA model to the time series
model = ARIMA(air, order=(1, 1, 1))
results = model.fit()

# Print the summary of the model's parameters and performance
print(results.summary())

# Plot the predicted values against the actual values
plt.plot(air)
plt.plot(results.predict(start='1949-02-01', end='1960-12-01'), color='red')
plt.xlabel('Year')
plt.ylabel('Passengers (in thousands)')
plt.title('Air Passengers Dataset')
plt.show()
```
