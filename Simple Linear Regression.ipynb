```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Define the data
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([3, 5, 7, 9, 11, 13])

# Fit linear regression model
model = LinearRegression().fit(x.reshape(-1, 1), y)

# Calculate estimated coefficient
coef = model.coef_[0]

# Make predictions using the model
y_pred = model.predict(x.reshape(-1, 1))

# Plot the data and regression line
plt.scatter(x, y, s=10)
plt.plot(x, y_pred, color='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

print(f"Estimated coefficient: {coef}")
```

In this example, we calculate the estimated coefficient by accessing the coef_ attribute of the LinearRegression object, which returns an array of coefficients. Since we only have one input variable, we take the first element of this array (coef_[0]). We then print the estimated coefficient to the console using an f-string.

The resulting output should show the estimated coefficient, which represents the slope of the regression line.
